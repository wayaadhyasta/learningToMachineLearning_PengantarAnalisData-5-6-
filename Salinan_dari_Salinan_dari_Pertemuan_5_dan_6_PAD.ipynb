```python
missing = ['na', '--']
df = pd.read_csv('/content/Apartemen.csv', na_values = missing)
```

```python
#Baca data kolom st milik
df['St_Milik']
```

```python
df['St_Milik'].isnull()
```

```python

cnt = 0

for row in df['St_Milik']:
  try:
    df['St_Milik'].isnull()
    int(row) #Mendeteksi angka pada kolom st milik
    df.loc[cnt, 'St_Milik'] = np.nan #apabila int maka akan disimpan pada variabel np.nan
  except ValueError:
    pass
  cnt+=1
```

```python
df['St_Milik'].isnull() #isnull untuk memastikan data ada missing value atau tidak
```

# Scalling


```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from pandas import DataFrame
import pandas as pd

df = pd.read_csv("/content/Apartemen_numerik.csv")
print(df)
```

```python
print(df.describe())
```

```python
df.loc[2,'Jum_kamar'] = 100
print(df)
print(df.shape)
print(df.Jum_kamar)
```

```python
fitur_minmax = df.copy()
Kol_minmax =['Jum_kamar']
features = fitur_minmax[Kol_minmax]
min_max = MinMaxScaler()
kamar_minmax=min_max.fit_transform(features.values)

print(kamar_minmax)
```

```python
fitur_minmax[Kol_minmax] = kamar_minmax
print(fitur_minmax)

df.to_csv('/content/Apartemen_minmax.csv', header=True, index=False)
```

```python
for col in df.columns.values:
  features = df[[col]]
  features_min_max = min_max.fit_transform(features)
  df[col] = features_min_max

df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
fitur_standar = df.copy()[['KodeApt']]
standarisasi = scaler.fit_transform(fitur_standar)

print(standarisasi)
```
